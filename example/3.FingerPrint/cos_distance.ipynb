```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
step = None
distance_m = {}
distance_t = {}

temp = 2200
indices = [format(i+1, '02') for i in range(10)]

for idx in indices:
    path_idx = os.path.join(f"{temp}K", f"{idx}.txt")
    data = np.loadtxt(path_idx, skiprows=1)

    if step is None:
        step = data[:,0]
    
    distance_m[idx] = data[:,1]
    distance_t[idx] = data[:,2]

num_step = len(step)
num_data = len(indices)

concat_m = np.zeros((num_data, num_step))
concat_t = np.zeros((num_data, num_step))
for i, idx in enumerate(indices):
    concat_m[i,:] = distance_m[idx]
    concat_t[i,:] = distance_t[idx]

concat_m = concat_m.reshape(1,-1).squeeze()
concat_t = concat_t.reshape(1,-1).squeeze()

x = np.arange(num_step*num_data)

plt.figure(figsize=(15,5))
plt.plot(x, concat_m, 'm-', label='M')
plt.plot(x, concat_t, 'c-', label='T')
# plt.scatter(x, concat_m, s=10, facecolor='m')
# plt.scatter(x, concat_t, s=10, facecolor='c')

verticals = np.arange(num_step, num_step*num_data, num_step )
for x_ver in verticals:
    plt.axvline(x_ver, color='k', linestyle='--', linewidth=1)

plt.xlabel('Step', fontsize=13)
plt.ylabel('Cosin distance', fontsize=13)
plt.xlim([0, num_data*num_step])
plt.ylim([0.2, 0.5])

plt.legend(loc='center right', fontsize=13)
plt.show()

```
